```
def split(df, group):
  data = namedtuple('data', ['filename', 'object'])
  gb = df.groupby(group)
  return [data(filename, gb.get_group(x)) for filename, x in zip(gb.groups.keys(), gb.groups)]
files = split(edit_csv, 'filename')
print(len(files))
```

```
err_file = []
i = 0
for group in files:
  i += 1
  try:
    cv2.imread(group[0])
    print(f"Going-on---{i}")
  except cv.error as e:
    err_file.append(files[i])
    print("Reading Error")
  continue
```
